## `Encoding`

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def label_encode_column(data, column_name):
    label_encoder = LabelEncoder()
    # Thực hiện Label Encoding
    data[column_name] = label_encoder.fit_transform(data[column_name])
    # In ra các giá trị unique sau khi encode
    unique_values = data[column_name].unique()
    print(f'Các giá trị unique sau khi encode trong cột {column_name}: {unique_values}')
    return data

# Gọi hàm với cột "weather"
data = label_encode_column(data, 'weather')
data
```

# **`MODEL`**

```python
#Loại bỏ cột 'date' 
data = data.drop('date',axis=1)
```

```python
data
```

```python
x = data.drop('weather',axis=1) #chứa các cột ngoại trừ cột 'weather'
y = data['weather'] #chứa dữ liệu trong cột 'weather'.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0) #X_test và y_test sẽ chiếm 25% tổng số dữ liệu và sẽ là ngẫu nhiên
```

```python
#Kích thước của các tập dữ liệu sau khi chia
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
#tính toán mean (trung bình) và standard deviation (độ lệch chuẩn) của từng đặc trưng trong X_train, 
#sau đó chuẩn hóa dữ liệu sao cho mean của mỗi đặc trưng là 0 và độ lệch chuẩn là 1.

X_test = sc.transform(X_test) #sử dụng các giá trị mean và độ lệch chuẩn đã được tính từ X_train để chuẩn hóa X_test
```

```python
X_train
```

## **`Logistic Regression`**

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm1 = confusion_matrix(y_test, y_pred)
print(cm1)
```

```python
sns.heatmap(cm1,annot=True, fmt='d')
plt.show()
```

```python
acc1 = accuracy_score(y_test, y_pred)
print(f"Accuracy score: {acc1}")
```

## **`Training the K-NN model on the Training set`**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm2 = confusion_matrix(y_test, y_pred)
print(cm2)
```

```python
sns.heatmap(cm2,annot=True)
plt.show()
```

```python
acc2 = accuracy_score(y_test, y_pred)
print(f"Accuracy score: {acc2}")
```

## **`Training the Naive Bayes on the Training set`**

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm3 = confusion_matrix(y_test, y_pred)
print(cm3)
```

```python
sns.heatmap(cm3,annot=True)
plt.show()
```

```python
acc3 = accuracy_score(y_test, y_pred)
print(f"Accuracy score : {acc3}")
```

## **`Random Forest Classifier`**

```python
from sklearn.ensemble import RandomForestClassifier
forest= RandomForestClassifier(n_estimators =40, random_state = 0)
forest.fit(X_train,y_train)
RandomForestClassifier(n_estimators=40, random_state=0)
```

```python
y_pred = forest.predict(X_test)
```

```python
cm4 = confusion_matrix(y_test,y_pred)
print(cm4)
```

```python
sns.heatmap(cm4,annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
acc4 = forest.score(X_test,y_test)
print(acc4)
```

# **`Tạo visualization cho Accuracy của từng Model`**

```python
mylist=[]
mylist2=[]
mylist.append(acc1)
mylist2.append("Logistic Regression")
mylist.append(acc2)
mylist2.append("KNN")
mylist.append(acc3)
mylist2.append("Naive Bayes")
mylist.append(acc4)
mylist2.append("RF")
plt.rcParams['figure.figsize']=8,6
sns.set_style("darkgrid")
plt.figure(figsize=(22,8))
ax = sns.barplot(x=mylist2, y=mylist, palette = "mako", saturation =1.5)
plt.xlabel("Classification Models", fontsize = 20 )
plt.ylabel("Accuracy", fontsize = 20)
plt.title("Accuracy of different Classification Models", fontsize = 20)
plt.xticks(fontsize = 11, horizontalalignment = 'center', rotation = 8)
plt.yticks(fontsize = 13)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.2%}', (x + width/2, y + height*1.02), ha='center', fontsize = 'x-large')
plt.show()
```
